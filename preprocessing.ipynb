```python
import pandas as pd
```

```python
marketplace = pd.read_csv('C:/Users/kiril/marketplace-project/marketplace.csv')
marketplace.head(10)
```

```python
marketplace = marketplace.rename(columns={'Unnamed: 0': 'Индекс'})
marketplace = marketplace.rename(columns={'start_session': 'Начало_сессии'})
marketplace = marketplace.rename(columns={'end_session' : 'Конец_сессии'})
marketplace = marketplace.rename(columns={'user_id' : 'ID_пользователя'})
marketplace = marketplace.rename(columns={'sex' : 'Пол'})
marketplace = marketplace.rename(columns={'platform': 'Платформа'})
marketplace = marketplace.rename(columns={'category': 'Категория'})
marketplace = marketplace.rename(columns={'consultation': 'Консультация'})
marketplace = marketplace.rename(columns={'cons_duration': 'Длительность_консультации'})
marketplace = marketplace.rename(columns={'number_of_products': 'Количество_товара'})
marketplace = marketplace.rename(columns={'basket_amount': 'Сумма_корзины'})
```

```python
marketplace['Начало_сессии'] = pd.to_datetime(marketplace['Начало_сессии'])
marketplace['Конец_сессии'] = pd.to_datetime(marketplace['Конец_сессии'])
```

```python
marketplace['Пол'] = marketplace['Пол'].astype('category')
marketplace['Платформа'] = marketplace['Платформа'].astype('category')
marketplace['Категория'] = marketplace['Категория'].astype('category')
```

```python
marketplace['Индекс'] = marketplace['Индекс'].astype('int32')
marketplace['ID_пользователя'] = marketplace['ID_пользователя'].astype('int32')
marketplace['Длительность_консультации'] = marketplace['Длительность_консультации'].astype('int32')
marketplace['Количество_товара'] = marketplace['Количество_товара'].astype('int32')
marketplace['Сумма_корзины'] = marketplace['Сумма_корзины'].astype('int32')
```

```python
marketplace.info()
```

```python
marketplace.to_csv('C:/Users/kiril/marketplace-project/marketplace_preprocessing.csv', index=True)
```
